```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

#Load the dataset
digits = load_digits()

#Dataset information
print("Dataset Shape:", digits.data.shape)
print("Image Shape:", digits.images.shape)
print("Target Label:", digits.target_names)

#Display the first image
plt.figure(figsize=(4,4))
plt.imshow(digits.images[0],cmap='pink')
plt.title(f"Label: {digits.target[0]}")
plt.axis('off')
plt.show()
```

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

digits = load_digits()

fig, axes =plt.subplots(2,5, figsize=(10,5))

for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='gray')
    ax.set_title(f"Digit: {digits.target[i]}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

**Q2. Write a program to display (a) no. of rows & column. (b) column names & (c) data types**

```python
#Import Libraries
import pandas as pd
from sklearn.datasets import load_iris

#Load the iris dataset
iris=load_iris(as_frame=True)

#Convert to Pandas DataFrame
df=iris.frame

#Display the number of rows and columns
print("Number of Rows and Columns:")
print(df.shape)

#Display the column names
print("\nColumn Names:")
print(df.columns)

#Display the data types
print("\nData Types:")
print(df.dtypes)
```

**Q3. Write a program to count the number of samples available in iris dataset & digit dataset.**

```python
#import datasets
from sklearn.datasets import load_iris, load_digits

#Load the iris dataset
iris = load_iris()

#Load the Digits dataset
digits = load_digits()

#Display the number of samples
print("Number of samples in Iris Dataset:", len(iris.data))
print("Number of samples in Digits Dataset:", len(digits.data))
```
